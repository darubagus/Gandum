```python
#No 3
'''
Menentukan setiap kolom numerik berdistribusi normal atau tidak. Gunakan normality test yang dikaitkan dengan histogram plot.
'''
sumbuUtama = df["SumbuUtama"]
alpha = 0.05
stat, p = stats.normaltest(sumbuUtama)
print('alpha = %.3f' %(alpha))
print('stat = %.3f, p-value = %.3f' % (stat, p))
# null hypotesis : x comes from a normal distribution
if (p > alpha):
    print('Data is normally distributed')
else :
    print('Data is not normally distributed')

ax = sns.distplot(sumbuUtama)
```

```python
#No 6
'''
Test korelasi: tentukan apakah setiap kolom non-target berkorelasi dengan kolom target, dengan menggambarkan juga scatter plot nya. Gunakan correlation test.
'''
# Kolom target == kolom ke-12 == Kelas
```

```python
'''
Pearson correlation coefficient (-1 â‰¤ r â‰¤ 1)
Bila r = 1, maka kedua kolom yang dibandingkan memiliki korelasi positif sempurna, artinya jika nilai x membesar, maka nilai y juga membesar dan vice versa
Bila r > 0, maka kolom x dan y memiliki korelasi positif
Bila r = 0, maka kolom x dan y tidak memiliki korelasi
Bila r < 0, maka kolom x dan y memiliki korelasi negatif
Bila r = -1, maka maka kedua kolom yang dibandingkan memiliki korelasi negatif sempurna, artinya jika nilai x membesar, maka nilai y mengecil dan vice versa.

Klasifikasi Pearson correlation coefficient
Perfect --> r +- 1
High Degree --> +- 0.5 < r < +- 1
Moderate Degree --> +- 0.3 < r < +- 0.49
Low Degree --> r < +- 0.29
No Correlation --> r = 0
'''
```

```python
# Kolom Daerah : Kolom Kelas
r = np.corrcoef(df["Daerah"], df["Kelas"])
print("r = %.3f" %(r[0][1]))
sns.regplot(df["Daerah"], df["Kelas"])
```
